```python
import numpy as np;import pandas as pd;import matplotlib.pyplot as plt;import seaborn as sns;sns.set()

raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73], 
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}
scores=pd.DataFrame(raw_data)
```

```python
plt.style.use('fivethirtyeight')
plt.scatter(scores.preTestScore,scores.postTestScore,s=scores.age,c='red')
```

```python
plt.scatter(scores.preTestScore,scores.postTestScore,s=4*scores.postTestScore,c=scores.female,cmap='rainbow')
```
