```python
import numpy as np;import pandas as pd;import matplotlib.pyplot as plt;import seaborn as sns;sns.set()

chipotle=pd.read_csv(r'C:\Users\SONY\Desktop\NumPy\chipotle.tsv.txt',sep='\t')

chipotle.head()
```

```python
chipotle.shape
```

### Visualizar los 5 items más vendidos.

```python
h=chipotle.item_name.value_counts().head(5)
plt.figure(figsize=(10,4))
h.plot(kind='bar')
plt.ylabel('Cuenta');plt.title('Items más vendidos')
```

```python
chipotle.item_price=chipotle.item_price.apply(lambda x: float(x[1:-1]))
```

```python
price=chipotle.groupby('order_id').sum()
sns.scatterplot(price.item_price,price.quantity,color='green',s=100,alpha=0.5)
```
