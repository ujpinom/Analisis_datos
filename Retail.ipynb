```python
import numpy as np;import pandas as pd;import matplotlib.pyplot as plt;import seaborn as sns;sns.set(style='ticks')
retail=pd.read_csv(r'C:\Users\SONY\Desktop\NumPy\Online_Retail.csv',encoding = 'latin1')
```

```python
retail=retail.loc[retail.Quantity>0,:]
```

```python
plt.style.use('fivethirtyeight')
ff=retail.groupby('Country')['Quantity'].sum().sort_values(ascending=False).drop('United Kingdom').head(10)
ff.plot(kind='bar')
plt.ylabel('')
```

```python
retail['Ingresos']=retail.UnitPrice*retail.Quantity
plt.style.use('seaborn-whitegrid')
intervalo=np.arange(0,50)
retail['Intervalo_precio']=pd.cut(retail.UnitPrice,intervalo)

re1=retail.groupby('Intervalo_precio')['Ingresos'].sum()
re1.plot()
plt.xticks(np.arange(0,50,3),np.arange(0,50,3))
plt.yticks(np.array([500000,1000000,1500000,2000000,2500000,3000000]),['$0.5 M','$ 1M','$ 1.5M','$ 2M','$ 2.5M','$ 3M'])
plt.legend()
```
