```python
import numpy as np;import pandas as pd;import matplotlib.pyplot as plt;import seaborn as sns;sns.set()
from seaborn import load_dataset
url='https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv'
titanic=pd.read_csv(url)
```

```python
titanic.head()
```

```python
titanic.set_index('PassengerId',inplace=True)
```

# Mostrar la proporción de ambos sexos

```python
import matplotlib.pyplot as plt
size=titanic.groupby('Sex').size().values
(size/titanic.shape[0])*100
labels=['Mujeres','Hombres']
plt.pie(size, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90,explode = (0.15 , 0))
plt.axis('equal')
plt.title('Proporción_sexos')
```

# scatterplot entre edades y Fare,discriminado por clase

```python
sns.scatterplot(titanic.Age,titanic.Fare,hue='Pclass',data=titanic,alpha=1,cmap='rainbow')
```

# Distribucción de supervivencia

```python
sns.countplot('Survived',data=titanic,hue='Sex',palette='Set3')
plt.title('Distribucción Muertos/Sobrevivientes')
```

# Fare hist

```python
sns.distplot(titanic.Fare,kde=False,bins=range(0,600,10),color='red')
```

```python
sns.catplot(x="Pclass", hue="Sex", col="Survived",
                data=titanic, kind="count",
                height=4, aspect=.7,palette='Set2');
```
