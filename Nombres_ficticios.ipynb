```python
import numpy as np;import pandas as pd;
```

### Crear datos ficticios para llevar a cabo operaciones sobre ellos

```python

raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

### Llevar dichos datos ficticios a un DataFrame

```python
df1=pd.DataFrame(raw_data_1);
df2=pd.DataFrame(raw_data_2);
df3=pd.DataFrame(raw_data_3);
```

```python
all_data=(df1.append(df2)) ## El método append() une a traves del eje 0;es decir apilando un df sobre otro
all_data
```

```python
all_data_col=pd.concat([df1,df2],axis=1) ##Concatenando a traves del eje 1;
all_data_col
```

Usando el método pd.merge() para unir all_data y df3 sobre la columna 'subject_id'

```python
pd.merge(all_data,df3,on='subject_id')
```

```python
pd.merge(df1,df2,on='subject_id')
```
