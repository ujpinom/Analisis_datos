```python
from sklearn.datasets import fetch_lfw_people
from sklearn.decomposition import PCA;import numpy as np;import matplotlib.pyplot as plt;import seaborn as sns;sns.set()
faces=fetch_lfw_people(min_faces_per_person=60)
print(faces.keys())
print(faces.target_names)
pca=PCA()
pca.fit(faces.data)
#plt.plot(np.cumsum(pca.explained_variance_ratio_))


modelo=PCA(150)
modelo.fit(faces.data)
fig,ax=plt.subplots(3,8,figsize=(9,4))
fig.subplots_adjust(hspace=0.1, wspace=0.1)
for i, axis in enumerate(ax.flat):
    axis.set(xticks=[],yticks=[])
    axis.imshow(modelo.components_[i].reshape(62,47),cmap='bone')
    

    
    

```

```python
plt.plot(np.cumsum(modelo.explained_variance_ratio_),'--r')
plt.xlabel('Número de componentes');plt.ylabel('Varianza acumulada')


```

```python
from sklearn.datasets import fetch_lfw_people
from sklearn.decomposition import PCA;import numpy as np;import matplotlib.pyplot as plt;import seaborn as sns;sns.set()
faces=fetch_lfw_people(min_faces_per_person=60)
fig,ax=plt.subplots(2,10,figsize=(10,2.5),subplot_kw={'xticks':[],'yticks':[]},gridspec_kw=dict(hspace=.5,wspace=0.1))
ax[0,0].set_ylabel('n_componentes\nde n')
ax[1,0].set_ylabel('150_comp.\nde n')
pca=PCA(150).fit(faces.data)
componentes=pca.transform(faces.data)
proyeccion=pca.inverse_transform(componentes)

for i in range(10):
    ax[0,i].imshow(faces.data[i].reshape(62,47),cmap='binary_r')
    ax[1,i].imshow(proyeccion[i].reshape(62,47),cmap='binary_r')


```
