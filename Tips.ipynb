```python
import numpy as np;import pandas as pd;import seaborn as sns;import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style("white")
sns.distplot(tips.total_bill,kde=False).set(xlabel='Cuenta[$]',ylabel='Frecuencia',title='Cuenta total')
```

```python

sns.jointplot(tips.total_bill,tips.tip,color='red')
```

```python
sns.pairplot(tips)
```

```python
sns.FacetGrid(tips,col='day',hue='sex').map(plt.hist,'tip');plt.legend()
```

```python
sns.stripplot(x='total_bill',y='day',data=tips,hue='sex')
plt.legend(loc='center right',ncol=2)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='time')
```

```python
sns.FacetGrid(tips,col='time').map(plt.hist,'tip')
```

```python
sns.FacetGrid(tips,col='sex',hue='smoker').map(plt.scatter,'total_bill','tip',alpha=0.5)
plt.legend()
```
