```python
import numpy as np;import pandas as pd;import datetime
wind_stats=pd.read_csv(r'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data',
                      sep="\s+",parse_dates=[[0,1,2]])
wind_stats
```

Las fechas proporcionadas no son consitentes con el mundo real. Se debe modificar dichas fechas haciendo uso del método apply()

```python
def modificar_anio(x):
    if x.year>1989:
        anio=x.year-100
    else:
        anio=x.year
    return pd.to_datetime(datetime.date(anio,x.month,x.day))

wind_stats['Yr_Mo_Dy']=wind_stats['Yr_Mo_Dy'].apply(modificar_anio)
```

```python
wind_stats.set_index('Yr_Mo_Dy',inplace=True)
```

Nos gustaría descubrir cuántos valores nulos hay por cada locación.

```python
wind_stats.isnull().sum()
```

```python
wind_stats.notnull().sum().sum()  ##Numero de muestras no nulas recogidas durante el periodo de vigilancia para todas las 
#locaciones.
```

Nos gustaría calcular la media,maximo,mínimo y la desviación éstandar de la velocidad del viento 
para cada una de las locaciones durante todo el periodo de registro.

```python
loc_stats=pd.DataFrame()
loc_stats[['Media','Max','Min','std']]=wind_stats.aggregate(['mean','max','min','std'],axis=0).T
loc_stats
```

```python
wind_stats.loc[wind_stats.index.month==1].mean()  ##Media de los vientos registrados  en los meses de enero para todas las locaciones
```
